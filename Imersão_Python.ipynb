**Aula 2 - Importação**

```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/_Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_Total_de_acoes = pd.read_excel("/content/_Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel("/content/_Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_Ticker
```

```python
df_Gemini = pd.read_excel("/content/_Imersão Python - Tabela de ações.xlsx", sheet_name="Gemini")
df_Gemini
```

```python
df_principal = df_principal[["Ativo",	"Data",	"Último (R$)",	"Var. Dia (%)"]].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns = {"Último (R$)":"valor_final" , "Var. Dia (%)":"var_dia_pct" }).copy()
df_principal
```

```python
df_principal["Var_pct"] = df_principal["var_dia_pct"] / 100
df_principal["valor_inicial"] = df_principal["valor_final"] / (df_principal['Var_pct']+1)
df_principal
```

```python
df_principal=df_principal.merge(df_Total_de_acoes, left_on="Ativo", right_on="Código", how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal["Variacao_$"] = (df_principal["valor_final"] - df_principal["valor_inicial"]) * df_principal["Qtde. Teórica"]
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal ["Qtde. Teórica"] = df_principal ["Qtde. Teórica"].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns= {'Qtde. Teórica' : "qtd_teorica"}).copy()
df_principal
```

```python
df_principal["Resultado"] = df_principal['Variacao_$'].apply(lambda valor: "Subiu" if valor > 0 else "Desceu" if valor < 0 else "Estável")
df_principal
```

```python
df_principal=df_principal.merge(df_Ticker, left_on="Ativo", right_on="Ticker", how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_Gemini
```

```python
df_principal = df_principal.merge(df_Gemini , left_on="Nome", right_on="Nome da empresa", how='left')
df_principal = df_principal.drop(columns=["Nome da empresa"])
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: 'Maior do que 100 anos' if x > 100 else 'Menor do que 50 anos' if x < 50 else 'Entre 50 e 100')
df_principal
```

```python

maior_valor = df_principal['Variacao_$'].max()
menor_valor = df_principal['Variacao_$'].min()
media_total = df_principal['Variacao_$'].mean()

# Calcular a média de quem subiu e a média de quem desceu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_$'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_$'].mean()

print("Maior\tR$", "{:,.2f}".format(maior_valor))
print("Menor\tR$", "{:,.2f}".format(menor_valor))
print("Média\tR$", "{:,.2f}".format(media_total))
print("Média de quem subiu\tR$", "{:,.2f}".format(media_subiu))
print("Média de quem desceu\tR$", "{:,.2f}".format(media_desceu))
```

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_$'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_segmento = df_principal.groupby('Resultado')['Variacao_$'].sum().reset_index()
df_analise_segmento
```

```python
fig = px.bar(df_analise_segmento , x ="Resultado", y="Variacao_$", text="Variacao_$", title= "Variacao Reais por Resultado" )
fig.show()
```
